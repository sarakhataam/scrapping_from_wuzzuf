```python
csvfile = open ('all_webdeva.csv','w',encoding='utf-8')
writer=csv.writer(csvfile)
writer.writerow(['campany','location','duration'])
for i in range(55):
    if i == 0:
        website = requests.get("https://wuzzuf.net/search/jobs/?q=web+developer&a=hpb").text
    else:
        website = requests.get(f"https://wuzzuf.net/search/jobs/?a=hpb&q=web%20developer&start={i}").text
    website=BeautifulSoup(website,"lxml")
    jobs=website.find_all('div',class_="css-1gatmva e1v1l3u10")
    for job in jobs:
        campany = job.find('div',class_="css-d7j1kk").a.text
        location= job.find("span",class_="css-5wys0k").text
        duration=job.find("div",class_="css-1lh32fc").text
        writer.writerow([campany,location,duration])
        print(i)
    
csvfile.close()
```
